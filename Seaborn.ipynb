# Seaborn Practice Questions and Answer

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=sns.load_dataset("tips")
print(data.head(10))
```

## Relational Visualization (Scatterplot)

```python
#Create a scatterplot showing the relationship between total_bill (x-axis) and tip (y-axis).
#Color the points based on the smoker status, and use the 'rocket' color palette.
sns.scatterplot(data=data,x="total_bill",y="tip",hue="smoker",palette="rocket")
plt.show()
```

## Distribution Visualization (Histogram)

```python
#Create a histogram of the tip amount.Separate the distributions by sex. Display the Kernel 
#Density Estimate line (kde) and set the number of bins to 15.
sns.histplot(data=data,x="tip",hue="sex",kde=True,bins=15)
plt.show()
```

## Categorical Data Points (Swarmplot)

```python
#Visualize the individual data points of tip against the day using a swarmplot. Separate the 
#swarms for each sex, and ensure there's a slight horizontal separation between the sex groups.
sns.swarmplot(data=data,x="day",y="tip",hue="sex",dodge=True)
plt.show()
```

## Categorical Aggregation (Barplot)

```python
#Generate a barplot comparing the median total_bill (y-axis) for each time of day (x-axis).
#You'll need to specify the estimator for the median calculation.
sns.barplot(data=data,x="time",y="total_bill",estimator=np.median)
plt.show()
```

# Relational Line Plot

```python
#Create a lineplot showing the relationship between size and the average total_bill (y-axis). 
#Color the lines based on the smoker status.
sns.lineplot(data=data,x="size",y="total_bill",hue="smoker")
plt.show()
```

## Density Visualization (KDE Plot)

```python
#Generate a KDE plot showing the distribution of total_bill. Group the distributions by day 
#using the hue parameter, and use the multiple='stack' parameter so the areas add up without overlapping.
sns.kdeplot(data=data,x="total_bill",hue="day",multiple="stack")
plt.show()
```

```python
df=data.copy()
```

# All-Pairs Visualization

```python
#Create a full grid of plots showing the pairwise relationships between all numeric columns in the df DataFrame.
sns.pairplot(df)
plt.show()
```

## Categorical Scatter (Stripplot)

```python
#Create a stripplot of tip versus day. Ensure that the data points are slightly randomized along 
#the x-axis to prevent them from stacking perfectly on top of one another.
sns.stripplot(data=data,x="day",y="tip",jitter=0.2)
plt.show()
```

## Categorical Distribution (Boxplot)

```python
sns.boxplot(data=data,x="day",y="total_bill",hue="time")
plt.show()
```
